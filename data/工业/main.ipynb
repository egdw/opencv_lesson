```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("./data/img.png")
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray = cv.cvtColor(img,cv.COLOR_RGB2GRAY)
plt.imshow(gray)
```

```python
# sobel 
img_blur = cv.GaussianBlur(gray, (3,3), 0)
sobelxy = cv.Sobel(src=img_blur, ddepth=cv.CV_64F, dx=1, dy=1, ksize=3)
plt.imshow(sobelxy)
```

```python
cv.imshow("demo",sobelxy)
cv.waitKey(0)
```
