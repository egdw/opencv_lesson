```python
import cv2
import numpy as np
import matplotlib.pyplot as plt # matplotlib库用于绘图展示
# 魔法指令，直接展示图，Jupyter notebook特有
%matplotlib inline
```

```python
img = cv2.imread("logo.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gaussian = cv2.GaussianBlur(img,ksize=(5,5),sigmaX=3)
plt.imshow(gaussian)
```

```python
canny = cv2.Canny(gaussian,30,100)
plt.imshow(canny)
```

```python
import math

lines = cv2.HoughLinesP(canny,rho=1,theta=math.pi/180,threshold=5,
                        minLineLength=1,maxLineGap=1)
img2 = cv2.imread("logo.jpg")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
for line in lines:
    line = line[0]
    cv2.line(img2,(line[0],line[1]),(line[2],line[3]),(255,0,0),1)
plt.imshow(img2)
```
